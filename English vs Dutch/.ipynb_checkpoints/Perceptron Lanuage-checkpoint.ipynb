```python
import numpy as np
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


with open("dutch.txt") as f:
    dutchString = " ".join(f.read().splitlines())

with open("english.txt") as f:
    englishString = " ".join(f.read().splitlines())


with open("dutchTestSet.txt") as f:
    dutchTestStr = " ".join(f.read().splitlines())

with open("englishTestSet.txt") as f:
    englishTestStr = " ".join(f.read().splitlines())
```

```python
 # turn the english and dutch words into a dict that counts the amount of characters in each sentence

def countLetters(words):
    # returns an np array (length of 26) where each index refers to the letter count
    counts = dict()
    rc = np.zeros(26)
    for character in words.lower():
        if (97 <= ord(character) <= 122):
            i = ord(character) - ord('a')
            if (0 <= i <= 26):
                rc[i] += 1
    return rc
```

```python
# this will make an X for both english and dutch

def make_X_and_y(text, language):

    length = text.count('.') + 1
    
    # ENGLISH MEANS y[i] = 1!!!!!!
    if (language == "english"): y = np.ones(length)
    else: y = np.full(length, -1)

    testArray = list()

    i = 0
    X = np.zeros(length)
    for sentence in text.split('.'):
        temp = countLetters(sentence)
        testArray.append(temp)
        i += 1
        
    X = np.vstack(testArray)

    return X, y
```

```python
X_e, y_e = make_X_and_y(englishString, "english")

X_d, y_d = make_X_and_y(dutchString, "dutch")

X = np.concatenate((X_e, X_d))
y = np.concatenate((y_e, y_d))
```

```python
# for the TEST:

X_eT, y_eT = make_X_and_y(englishTestStr, "english")

X_dT, y_dT = make_X_and_y(dutchTestStr, "dutch")

X_test = np.concatenate((X_eT, X_dT))
y_test = np.concatenate((y_eT, y_dT))
```

```python
def unison_shuffled_copies(a, b):
    assert len(a) == len(b)
    p = np.random.permutation(len(a))
    return a[p], b[p]

X,  y = unison_shuffled_copies(X, y)
X_test, y_test = unison_shuffled_copies(X_test, y_test)
```

```python
def perceptronTrain(X, y, num_epochs):
  weights = np.zeros(X.shape[1])
  bias = 0

  # loop for maxIter
  for epoch in range(num_epochs):
    for i in range(X.shape[0]):
      # print(f"epoch {epoch}, looking at instance {i}")
      total = weights.dot(X[i]) + bias
      if (total * y[i]) <= 0:
        # if mistake is made: update weights and bias
        weights = weights + (X[i] * y[i])
        bias += y[i]


  # loop over training data

  return weights, bias
```

```python
weights, bias = perceptronTrain(X, y, 20)
```

```python
def classify(weights, bias, example):
    activation = example.dot(weights) + bias
    if not activation: return 0
    return 1 if activation > 0 else -1
```

```python
def runTests(wieghts, bias, X_test, y_test):
  total_correct = 0
  for i in range(X_test.shape[0]):
    predicted = classify(wieghts, bias, X_test[i])
    if (predicted == y_test[i]):
      total_correct += 1

  accuracy = total_correct / len(y_test)
  return accuracy
  # print(f"accuracy: {accuracy}")
```
