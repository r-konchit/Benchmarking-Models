```python
import numpy as np

from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
spambase = fetch_ucirepo(id=94) 
  
# data (as pandas dataframes) 
X = spambase.data.features.to_numpy()
y = spambase.data.targets.to_numpy().flatten()



from sklearn.model_selection import train_test_split
```

```python
# return a trained perceptron model
# weights, bias

def perceptronTrain(X, y, num_epochs):
  weights = np.zeros(X.shape[1])
  bias = 0

  # loop for maxIter
  for epoch in range(num_epochs):
    for i in range(X.shape[0]):
      # print(f"epoch {epoch}, looking at instance {i}")
      total = weights.dot(X[i]) + bias
      if (total * y[i]) <= 0:
        # if mistake is made: update weights and bias
        weights = weights + (X[i] * y[i])
        bias += y[i]


  # loop over training data

  return weights, bias
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
weights, bias = perceptronTrain(X_train, y_train, 20)
```

```python
def classify(weights, bias, example):
    activation = example.dot(weights) + bias
    if not activation: return 0
    return 1 if activation > 0 else -1
```

```python
classify(weights, bias, X[55])
```

```python
def runTests(wieghts, bias, X_test, y_test):
  total_correct = 0
  for i in range(X_test.shape[0]):
    predicted = classify(wieghts, bias, X_test[i])
    if (predicted == y_test[i]):
      total_correct += 1

  accuracy = total_correct / len(y_test)
  print(f"accuracy: {accuracy}")
```

```python
runTests(weights, bias, X_test, y_test)
```
